# Linear Regression with NumPy

In this assignment you will implement linear regression **from scratch using only NumPy** — no sklearn, no autograd.

You will cover:

- **Closed-form solution**: Derive and implement the Normal Equation $\hat{w} = (X^\top X)^{-1} X^\top y$ to find the optimal weights analytically.
- **Gradient descent**: Implement batch gradient descent by computing the MSE loss and its gradient with respect to the weights at each iteration.
- **Evaluation**: Measure model quality using Mean Squared Error (MSE) and $R^2$ score on a held-out test set.
- **Visualization**: Plot the fitted regression line against the data and the loss curve over training epochs.
